```python
import random
```

```python
#Implementação da parte 1 do Exercício
def mdc(valorA : int, valorB : int, shouldPrint : bool = False) -> int:
    dividendo : int = max(valorA, valorB)
    divisor : int = min(valorA, valorB)
    resto : int = dividendo % divisor
    
    if(shouldPrint):
        quociente : int = dividendo / divisor
        print(f"dividendo: {dividendo}, divisor: {divisor}, resto: {resto}, quociente: {quociente}")
        
    while(resto != 0):
        dividendo = divisor
        divisor = resto
        resto = dividendo % divisor
        quociente = dividendo / divisor
        
        if(shouldPrint):
            quociente : int = dividendo / divisor
            print(f"dividendo: {dividendo}, divisor: {divisor}, resto: {resto}, quociente: {quociente}")
    
    return divisor
```

```python
#Teste 1 da implementação da parte 1 do Exercício 9
print("Informe o valor de A:")
valorA : int = int(input())

print("Informe o valor de B")
valorB : int = int(input())

print(f"--- Calculando MDC({valorA}, {valorB})")

resultadoMdc: int = mdc(valorA, valorB, True)

print(f"MDC({valorA}, {valorB}) = {resultadoMdc}")
```

```python
#Teste 2 da implementação da parte 1 do Exercício 9
print("Informe a quantidade de números aleatórios que deseja testar:")
quantidadeNumeros : int = int(input())

quantidadeResultados1 : int = 0;

for i in range(0, quantidadeNumeros):
    valorA : int = random.randint(1, 10000)
    valorB : int = random.randint(1, 10000)
        
    resultadoMdc : int = mdc(valorA, valorB)
    print(f"---- Resultado MDC({valorA}, {valorB}) = {resultadoMdc}")
    
    if(resultadoMdc == 1):
        quantidadeResultados1 = quantidadeResultados1 + 1

proporcaoResultados1 = quantidadeResultados1 / quantidadeNumeros
print(f"Proporção de resultados 1 foi: ({quantidadeResultados1} / {quantidadeNumeros}) = {proporcaoResultados1}")
```
